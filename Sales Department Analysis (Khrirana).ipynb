##### a.Melakukan uji stasioner menggunakan ADF
  Hal yang perlu diperhatikan dalam uji stasioner 


1.   Nilai ADF Statistic harus lebih besar dari nilai critical value
2.   Nilai p-value > 0.05
> Dengan uji hipotesis :
  *   H0 : Data tidak bersifat stasioner
  *   H1 : Data bersifat stasioner


```python
# stasionarity check
from statsmodels.tsa.stattools  import adfuller
sales_adf = adfuller(df_sales_analysis['sale_price'], autolag= 'AIC')
sales_adf
```

```python
# Supaya hasil uji lebih nyaman dibaca, maka kita buat series berdasarkan index 
dfoutput=pd.Series(sales_adf[0:4], index=['Test Statistic','p-value','Lags Used','No. of Obs'])

# Gunakan for loop untuk menaruh value dari dictionary dengan function items()
for key,value in sales_adf[4].items():
    dfoutput['Critical Value (%s)'%key] = value

print('Dickey-Fuller Test: ')
print(dfoutput)

# test statistics = ADF test
```

> insight = 

```python
# Plot dalam bentuk visualisasi line chart dengan menggunakan function numpy np.log
plt.figure(figsize=(20,10))
sales_adf_log=np.log(df_sales_analysis) 
plt.plot(sales_adf_log)
```

```python
# Cek secara visual kembali dengan plot data_log dengan rolling window mean dan std. deviation
plt.figure(figsize=(20,10))

MAvg=sales_adf_log.rolling(window=12).mean() 
MStd=sales_adf_log.rolling(window=12).std()

plt.plot(sales_adf_log)
plt.plot(MAvg, color='orange')
plt.plot(MStd, color='red')
```

```python
# Hitung difference dari data_log dengan MAvg
sales_log_diff=sales_adf_log-MAvg

# Tampilkan 12 baris teratas dari data_log_diff
sales_log_diff.head(12)
```

```python
sales_log_diff = sales_log_diff.dropna()
sales_log_diff.head()
```

```python
# Buat function stationarity untuk menampilkan multiple line chart dari actual, rolling mean, dan rolling std.deviasi
# yang juga berisikan hasil tes dari Dickey-Fuller

def stationarity(timeseries):
    
    rolmean=timeseries.rolling(window=12).mean()
    rolstd=timeseries.rolling(window=12).std()
    
    plt.figure(figsize=(20,10))
    actual=plt.plot(timeseries, color='green', label='Actual')
    mean_6=plt.plot(rolmean, color='blue', label='Rolling Mean') 
    std_6=plt.plot(rolstd, color='black', label='Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show(block=False)
    
    print('Dickey-Fuller Test: ')
    dftest=adfuller(timeseries['sale_price'], autolag='AIC')
    dfoutput=pd.Series(dftest[0:4], index=['Test Statistic','p-value','Lags Used','No. of Obs'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print(dfoutput)
```

```python
# panggil function dan terapkan pada data_log_diff
stationarity(sales_log_diff)
```

```python
# Buat format ukuran chart
plt.figure(figsize=(20,10))

# Terapkan shift terhadap data yang telah diterapkan transformasi log di cell sebelumnya
sales_shift=sales_adf_log-sales_adf_log.shift() 

# Tampilkan chart
plt.plot(sales_shift)
```

```python
sales_shift = sales_shift.dropna()
stationarity(sales_shift)
```

> insight :  Jika dilihat dari data yang mengggunakan status complete, data tidak bersifat stasioner ketika dilakukan uji ADF dan dilakukan differensiasi untuk menstabilkan nilai sale_price.
> **Dari hasil akhir terlihat nilai Test Statistics >  Critical Values, yang artinya kita gagal menolak hipotesis null, maka data bersidat non-stasioner.**
